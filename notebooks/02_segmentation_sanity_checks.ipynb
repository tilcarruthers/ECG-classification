# 02 Segmentation Sanity Checks

Use this notebook to verify that beat markers and extracted windows are aligned sensibly.

## Purpose

Compare segmentation methods before large retrains:

- `notebook_style`
- `aligned_interval`
- `centered_window`

```python
import numpy as np
from datasets import load_dataset
from ecg_repo.data.loading import get_records_list, infer_record_keys
from ecg_repo.data.segmentation import get_segments

dataset = load_dataset('dpelacani/ecg-led2-cleaned')
records = get_records_list(dataset, split='train')
keys = infer_record_keys(records[0])
record = records[0]
signal = np.asarray(record[keys['signal']], dtype=float)
beat_locs = np.asarray(record[keys['beat_locs']], dtype=int)

methods = ['notebook_style', 'aligned_interval', 'centered_window']
for method in methods:
    segments = get_segments(signal, beat_locs, method=method)
    print(method, len(segments), segments[:3])
```

## What to look for

- obvious off-by-one alignment
- whether segments contain the beat they are labelled as
- whether final windows truncate too aggressively
- whether interval-based segmentation is too dependent on heart-rate variation
