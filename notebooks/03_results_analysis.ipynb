# 03 Results Analysis

Use this notebook only after training runs are complete.

## Purpose

Load saved metrics, predictions, and plots from `outputs/runs/*`.
Do not retrain models here.

```python
from pathlib import Path
import json
import pandas as pd

runs_root = Path('../outputs/runs')
run_dirs = sorted([p for p in runs_root.glob('*') if p.is_dir()])
summary_rows = []
for run_dir in run_dirs:
    metrics_path = run_dir / 'artifacts' / 'test_metrics.json'
    if metrics_path.exists():
        metrics = json.loads(metrics_path.read_text())
        summary_rows.append({
            'run': run_dir.name,
            'test_accuracy': metrics['multiclass']['accuracy'],
            'test_macro_f1': metrics['multiclass']['macro_f1'],
            'test_balanced_accuracy': metrics['multiclass']['balanced_accuracy'],
        })
summary = pd.DataFrame(summary_rows).sort_values('test_macro_f1', ascending=False) if summary_rows else pd.DataFrame()
summary
```
